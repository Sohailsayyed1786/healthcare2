# Data Splitting into training and testing part

```python
from sklearn.model_selection import train_test_split

# All features
features_all = df.drop(columns=['DEATH_EVENT'])
target = df['DEATH_EVENT']

# Serum creatinine and ejection fraction only
features_creatinine_ejection = df[['serum_creatinine', 'ejection_fraction']]

# Split the data into training and testing sets (80-20 split)
X_train_all, X_test_all, y_train, y_test = train_test_split(features_all, target, test_size=0.2, random_state=42)
X_train_creatinine_ejection, X_test_creatinine_ejection = train_test_split(features_creatinine_ejection, test_size=0.2, random_state=42)
```

# Model_building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return accuracy, precision, recall, f1
```

# List_Of_Models_Evaluate

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    
    'Decision Tree': DecisionTreeClassifier(),
    
    'Random Forest': RandomForestClassifier(),
    
    'Support Vector Machine': SVC(probability=True),
    
    'K-Nearest Neighbors': KNeighborsClassifier(),
    
    'Naive Bayes': GaussianNB(),
    
    'Gradient Boosting': GradientBoostingClassifier(),
    
    'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

# Evalute_the_models_with_All_Features

```python
print("Evaluation with all features:")
for name, model in models.items():
    accuracy, precision, recall, f1= evaluate_model(model, X_train_all, y_train, X_test_all, y_test)
    print(f"{name}: Accuracy={accuracy:.4f}, Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}")
```

# Evalute_the_models_with_serum creatinine and ejection fraction

```python
print("\nEvaluation with serum creatinine and ejection fraction:")
for name, model in models.items():
    accuracy, precision, recall, f1= evaluate_model(model, X_train_creatinine_ejection, y_train, X_test_creatinine_ejection, y_test)
    print(f"{name}: Accuracy={accuracy:.4f}, Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}")
```
